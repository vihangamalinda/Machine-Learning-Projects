```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("DataSet/spambase_raw.csv",header = None)
df.head()
```

```python
df.columns =["X"+str(i) for i in range(0,len(df.columns)-1)]+["y"]
#converting the class type column to y[57] while others are converted as X+index(independant variables)
df.head()
```

```python
X_trn,X_tst,y_trn,y_tst = train_test_split(
    df.drop("y",axis=1),
    df.y,
    test_size =0.3,
    random_state =1
)

#combing the X data with y data on train and test
trn = X_trn
trn["y"] = y_trn

tst = X_tst
tst["y"] = y_tst
```

```python
sns.countplot(x="y",data =df)
```

```python
g = sns.PairGrid(
    trn,
    vars= ["X5","X20","X25","X53"],
    hue="y",
    diag_sharey = False,
    palette =["red","green"]
)

#ploting a histrogram on the diaganol
g.map_diag(plt.hist)
#plotting a density plot above the diaganol
g.map_upper(sns.kdeplot)
#plotting a scatter plot below the diagnol
g.map_lower(sns.scatterplot)

g.add_legend(title="Spam")
```

```python
#As you can see there is a clear variance in the data set
df.to_csv("DataSet/SpamBase.csv",sep=",",index=False)
trn.to_csv("DataSet/train_SpamBase.csv",sep=",",index=False)
tst.to_csv("DataSet/test_SpamBase.csv",sep=",",index=False)
```
