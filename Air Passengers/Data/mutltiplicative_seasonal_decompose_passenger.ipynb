```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.dates import DateFormatter
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("DataSet/AirPassengers.csv",parse_dates=["Month"],index_col=["Month"])
df.head()
```

```python
df.describe()
```

```python
fig,ax = plt.subplots()
plt.xlabel("Year:1949-1960")
plt.ylabel("Monthly Passengers(1000s)")
plt.title("Monthly Intl Air Passengers")
plt.plot(df,color="black")
ax.xaxis.set_major_formatter(DateFormatter("%Y"))
#shows a upward trend with a sesonal trend
```

```python
#fig size setup
plt.rcParams["figure.figsize"]=[7,8]

#period(how many observations are there before repeating) argument given 12 as like months
#plotting the seasonal decomposition
sd = seasonal_decompose(df,period=12).plot()

```

```python
#above residual grap shows that its better to use multiplicative model more than addative model
sd= seasonal_decompose(df,model="multiplicative").plot()
```
