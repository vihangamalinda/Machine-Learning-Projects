```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
trn = pd.read_csv("datasets/optdigits_trn.csv")
X_trn = trn.drop("y",axis=1)
y_trn = trn.y

tst = pd.read_csv("datasets/optdigits_tst.csv")
X_tst = tst.drop("y",axis = 1)
y_tst = tst.y
```

```python
#setting up the Principle component analysis object
pca = PCA()

#transforming the training dataset trn_tf
trn_tf = pca.fit_transform(X_trn)

plt.plot(pca.explained_variance_ratio_)
#Large amount of variance can be seen in one or two variable in the data set.
#Because of that dimentionality reduction will be successful on the data set
```

```python
# Projecting the first two principle(PCA1,PCA2) componets on a plot and give colours according to its class
# trn_tf[:,0]= all the rows of the data sets in the PCA1 
# trn_tf[:,1]= all the rows of the data sets in the PCA2
sns.scatterplot(
    x= trn_tf[:,0],
    y= trn_tf[:,1],
    style= y_trn, #Grouping variable that will produce points with different markers
    hue= y_trn, #Grouping variable that will produce points with different colors
    palette = ["red","green","blue"]
)
```

```python
#reflection on original data set on pca
print("%.2f"%pca.score(X_trn))
```

```python
#transform testing dataset to tst_tf
tst_tf = pca.fit_transform(X_tst)

sns.scatterplot(
    x = tst_tf[:,0],
    y = tst_tf[:,1],
    style =y_tst,
    hue = y_tst,
    palette =["red","blue","green"]
)
```

```python
print("%.2f"%pca.score(X_tst))
```
