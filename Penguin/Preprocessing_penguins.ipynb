```python
from palmerpenguins import load_penguins
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
df = load_penguins()

#df.isnull().sum()
#species               0
#island                0
#bill_length_mm        2
#bill_depth_mm         2
#flipper_length_mm     2
#body_mass_g           2
#sex                  11
#year                  0

#droping the year ,sex and island colum and null values
df = df.drop(["year","sex","island"],axis=1).dropna()
#rename species to y
df = df.rename(columns={"species":"y"})
df.head()
```

```python
#visualizing the data set according to their spicies(y)
sns.countplot(x="y",data=df)
```

```python
#making grid using seabon
g = sns.PairGrid(
    df,
    vars= ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"],
    hue="y",
    diag_sharey = False,
    palette =["red","green","blue"]
)

#ploting a histrogram on the diaganol
g.map_diag(plt.hist)
#plotting a density plot above the diaganol
g.map_upper(sns.kdeplot)
#plotting a scatter plot below the diagnol
g.map_lower(sns.scatterplot)

g.add_legend()
```

```python
df.to_csv("DataSet/penguins.csv",sep=",",index= False)
```
