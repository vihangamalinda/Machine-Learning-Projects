```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("DataSet/penguins.csv")

y= df.y
df =df.drop("y",axis=1)

#has to use standard scaler to sandardize data frame because data are in differnet scale
df = pd.DataFrame(
    StandardScaler().fit_transform(df),
    columns = df.columns
)

df.head()
```

```python
#use coustom score where it use silhouette score
def s2(estimator, X):
    return silhouette_score(X,estimator.predict(X))

#cluster range 
param = range(2,10)

km =KMeans(random_state=0, init = "k-means++")

#grid search cv object and set its variables
grid = GridSearchCV(
    km,
    {"n_clusters":param},
    scoring = s2,
    cv=2)

grid.fit(df)

#selecting the best model from the gridSearch cv
best_km = grid.best_estimator_

#shwoing the optimatal model through best_km
best_km.get_params()

```

```python
plt.plot(
    param,
    grid.cv_results_["mean_test_score"]
)

plt.axvline(
    x= best_km.n_clusters,
    color = "red",
    ls= "--"    
)

plt.xlabel("Total Centers")
plt.ylabel("Sihouette Score")
```

```python
sns.scatterplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data =df,
    hue = y,
    style = best_km.labels_,
    palette = ["orange","blue","green"]
)

plt.scatter(
    best_km.cluster_centers_[:,0],
    best_km.cluster_centers_[:,1],
    marker="x",
    s=200,
    c="red"
    
)
```
